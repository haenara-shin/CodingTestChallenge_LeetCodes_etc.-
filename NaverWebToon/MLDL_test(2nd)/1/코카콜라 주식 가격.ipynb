### Data

주어진 데이터는 다음과 같습니다.

##### coca-cola.csv: 연도별 코카콜라의 주당 가격을 담은 파일

* Year: 연도
* Average Stock Price: 평균 주당 가격(USD/1주)

* Average USD/KRW: 평균 환율

##### 3. 연도별 코카콜라 사의 주당 가격을 원화로 구하기

```python
# 코드를 작성해주세요
df3['Average Stock Price(KRW)'] = round(df3['Average USD/KRW'] * df3['Average Stock Price'], 2)
df3
```
